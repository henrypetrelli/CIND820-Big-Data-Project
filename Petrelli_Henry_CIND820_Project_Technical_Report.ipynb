Spatial Analysis of Violent Crime Hotspots in Chicago (2005–2025)

Data Acquisition:

The analysis begins by collecting crime data from the City of Chicago’s public data API. Because the dataset is large, the data is retrieved in chunks using a loop with an offset parameter. This ensures that all records for each selected year are downloaded without exceeding API limits.

The query is structured to only retrieve relevant columns, including:

Unique identifier (id)
Date of occurrence
Crime type (primary_type)
Geographic coordinates (latitude and longitude)

The process is repeated for three specific years: 2005, 2015, and 2025. These years are chosen to represent long-term temporal changes. 
Each chunk of data is appended to a list, and all chunks are later combined into a single dataset.


Data Cleaning and Preparation:

After the raw dataset is assembled, several preprocessing steps are applied to ensure data quality and consistency.

Column names are standardized to lowercase and stripped of whitespace to avoid referencing issues. Latitude and longitude values are converted into numeric format, and invalid entries are coerced into missing values.

The date column is converted into a datetime format, allowing extraction of the year for filtering and grouping.

Rows with missing or invalid values in critical fields (date, latitude, longitude) are removed. The dataset is also filtered to include only records that fall within the geographic boundaries of Chicago, ensuring spatial accuracy.

Duplicate records are removed using the unique identifier to avoid counting the same crime multiple times.

Crime type labels are standardized to uppercase and cleaned of whitespace. The dataset is then filtered to include only violent crimes commonly associated with gang activity:

Homicide
Assault
Battery
Robbery
Weapons violations

Only relevant columns are retained, and the dataset index is reset.


Descriptive Data Analysis:

Initial exploration of the dataset is conducted to understand its structure and distribution.

Descriptive statistics are computed for geographic coordinates, including mean, median, and range. These values confirm that the data is spatially concentrated within a reasonable geographic area.

The most frequent crime type is identified using the mode, which reveals that battery is the most common offense. A frequency count of crime types confirms the dominance of battery and assault.

This step ensures that the dataset is clean, consistent, and suitable for spatial analysis.


National (FBI) Crime Data Integration:

The FBI’s National Crime Trends (2005 - 2024) containing national crime statistics is imported. This dataset includes annual violent crime rates and total crime rates.

Only relevant columns are retained to simplify analysis. Descriptive statistics are calculated to understand overall trends, including mean, median, and range.

This dataset is used later to compare local hotspot trends with national crime trends.


Crime Trend Visualization:

Crime trends are visualized by grouping incidents by year and crime type. The most common crime types are identified and plotted over time.

The visualization shows how crime counts change across the selected years. Labels are added to each data point for clarity, with adjustments made to prevent overlapping text.

A second visualization compares violent and non-violent crime rates at the national level using a stacked bar chart. This helps illustrate the proportion of violent crime relative to total crime.

These visualizations provide a high-level understanding of how crime patterns evolve over time.


DBSCAN Model Developement: 

K-nearest approach:

To identify spatial crime hotspots, the DBSCAN clustering algorithm is applied.

The dataset is filtered again to ensure only relevant violent crimes are included. Geographic coordinates are extracted and converted into radians, which is necessary for accurate distance calculations using the haversine metric.

A k-nearest neighbors approach is used to construct a k-distance graph. This graph helps determine an appropriate value for the epsilon (eps) parameter, which defines the neighborhood radius for clustering.

Parameter Tuning:

Multiple eps values are tested to evaluate how clustering behavior changes. For each value, DBSCAN is applied separately to each year.

The number of clusters and noise points is recorded for each configuration. Smaller eps values produce many small clusters with more noise, while larger values merge clusters and reduce noise.

An optimal epsilon value is selected based on achieving a balance between meaningful cluster formation and minimal noise.


Running the DBSCAN Model:

The developed DBSCAN model is applied using the selected parameters.

Clustering is performed separately for each year to avoid mixing spatial patterns across time. Each crime is assigned a cluster label, where -1 represents noise.

The number of clusters and noise points is reported for each year. Results show an increase in the number of clusters over time, suggesting that hotspots are becoming more fragmented.


Stability Analysis:

To evaluate the robustness of the clustering results, a stability-based validation approach is used.

The dataset is randomly split into training and testing subsets multiple times. DBSCAN is applied to the training data, and the distance between test points and the nearest cluster cores is calculated.

The average distance across iterations is used as a stability score. This value is converted from radians to kilometers for interpretability.

A very small average distance indicates that clusters are highly stable and consistent across different samples of the data.


Hotspot Visualization:

Clustered data is visualized using scatter plots. Points belonging to clusters are colored based on their cluster label, while noise points are shown in gray.

Cluster centroids are calculated and labeled on the plot, allowing for easy identification of hotspot locations.

These visualizations provide a clear representation of how crime is spatially distributed in each year.


Hotspot Movement Analysis:

To analyze how hotspots evolve over time, cluster centroids are compared across years.

Pairwise distances between centroids are computed, and each cluster is matched with the closest cluster in the next time period.

Clusters are classified based on their movement:

Persistent clusters show minimal movement
Slightly shifted clusters show moderate movement
Different clusters show significant relocation

This allows for tracking the spatial evolution of crime hotspots over time.

Cluster Lineage Construction:

A lineage mapping is created to track clusters across all three years.

Each cluster from 2005 is linked to its closest counterpart in 2015, and then to 2025. Movement distances are recorded for each transition.

Clusters are classified as:

Persistent if movement is minimal across both intervals
Shifting if movement exceeds a defined threshold
Disappeared if no corresponding cluster exists

This provides a structured view of how hotspots persist or change over time.


Stability Ranking:

Clusters are ranked based on total movement across both time intervals.

Clusters with the smallest total movement are considered the most stable. These represent long-term persistent hotspots.

Most clusters show some degree of movement, indicating that while crime remains concentrated, its exact location may shift slightly over time.


Model Comparison:

DBSCAN is compared with alternative clustering algorithms, including K-Means and Agglomerative clustering.

A subset of the data is used to reduce computational complexity. Each model is evaluated using a performance metric, such as the silhouette score.

For DBSCAN, the stability score calcuated earlier from the stability analysis is used for comparison rather than a silhouette score. This score is inverted so that it can be plotted on the same graph as the silhouette scores of the other clustering models with a normalized scale raning from 0-1.

While K-Means performs reasonably well, it requires specifying the number of clusters in advance. Agglomerative clustering performs slightly worse but similarly overall.

DBSCAN is preferred because it better captures the natural structure of spatial data and effectively handles noise. This is reflected in the comparison plot as DBSCAN has a much higher score than the other two models. 


Chicago vs National Comparison:

The strength of crime hotspots in Chicago is calculated as the proportion of points assigned to clusters.

This metric is compared with national violent crime rates for the same time periods.

Results show that while national violent crime rates have declined significantly, the strength of Chicago hotspots remains nearly constant and very high.
